```python
class Livros():
    
    # Esse método inicializa cada objeto criado a partir dessa classe
    def __init__(self, title: str, vol: int, writer: Any, ed: str, year: int, editor: str, isbn: int, page: int, gender: Any, lang_o: str, lang: str, translate: Any, cover: str, fornec: str, buyprice: float, profit : float, estoque: int) -> None: 
        
        # Aqui vem os atributos de cada objeto que é criado a partir dessa classe
        self.title = title
        self.vol = vol
        self.__writer = writer
        a = 0
        t = len(self.__writer)
        if type(self.__writer) == type("string"):
            self.author = writer
        else:
            self.author = ""
            while a < t-1:
                self.author += self.__writer[a] + ", "
                a += 1
            self.author += "e " + self.__writer[t-1]
        self.ed = ed
        self.year = year
        self.editor = editor
        self.isbn = isbn
        self.page = page  
        self.__gender = gender
        b = 0
        u = len(self.__gender)
        if type(self.__gender) == type("string"):
            self.genders = gender
        else:
            self.genders = ""
            while b < u-1:
                self.genders += self.__gender[b] + ", "
                b += 1
            self.genders += "e " + self.__gender[u-1]
        self.lang_o = lang_o
        self.lang = lang
        self.__translate = translate
        c = 0
        v = len(self.__translate)
        if type(self.__translate) == type("string"):
            self.translator = translate
        else:
            self.translator = ""
            while c < v-1:
                self.translator += self.__translate[c] + ", "
                c += 1
            self.translator += "e " + self.__translate[v-1] 
        self.cover = cover
        self.valor = buyprice * profit
        self.__profit = profit
        self.__buyprice = buyprice
        self.__fornec = fornec
        self.__estoque = estoque
    
    def __str__(self):
        return """O livro: %s, volume %s %s edição, foi escrito por %s no ano %s e foi publicado por %s e tem %s páginas.
O livro tem a capa %s, é do(s) gênero(s) %s e foi escrito originalmente em %s.""" % (self.title, self.vol, self.ed, self.author, self.year, self.editor, self.page, self.cover, self.genders, self.lang_o)
    
    # Imprime na tela informações sobre o livro
    def informar(self) -> None:
        print(f"Título: {self.title}")
        print(f"Volume: {self.vol}")
        print(f"Autor(a/es/as): {self.author}")
        print(f"Edição: {self.ed}")
        print(f"Ano: {self.year}")
        print(f"Editora: {self.editor}")
        print(f"ISBN: {self.isbn}")
        print(f"Número de páginas: {self.page}")
        print(f"Gênero(s): {self.genders}")
        print(f"Idioma Original: {self.lang_o}")
        print(f"Traduzido para: {self.lang}")
        print(f"Traduzido por: {self.translator}")
        print(f"Tipo de capa: {self.cover}")
        print(f"Preço: R$ {self.valor}")
        print(f"Quantidade em estoque: {self.__estoque}")
    
    def __infos(self) -> None:
        print(f"Título: {self.title}")
        print(f"Volume: {self.vol}")
        print(f"Autor(a/es/as): {self.author}")
        print(f"Edição: {self.ed}")
        print(f"Ano: {self.year}")
        print(f"Editora: {self.editor}")
        print(f"ISBN: {self.isbn}")
        print(f"Número de páginas: {self.page}")
        print(f"Gênero(s): {self.genders}")
        print(f"Idioma Original: {self.lang_o}")
        print(f"Traduzido para: {self.lang}")
        print(f"Traduzido por: {self.translator}")
        print(f"Tipo de capa: {self.cover}")
        print(f"Fornecedor: {self.__fornec}")
        print(f"Preço de compra: R$ {self.__buyprice}")
        print(f"Margem de lucro: {self.__profit} %")
        print(f"Preço de venda: R$ {self.valor}")
        print(f"Quantidade em estoque: {self.__estoque}")        
        
    def relatorio(self, tipo: str) -> None:
        tipo = tipo.lower()
        if tipo == "completo":
            return self.__infos()
        if tipo == "estoque":
            print(f"Quantidade em estoque: {self.__estoque}")
        if tipo == "prices":
            print(f"Preço de compra: R$ {self.__buyprice}")
            print(f"Margem de lucro: {self.__profit} %")
            print(f"Preço de venda: R$ {self.__valor}")
                     
```

```python
class Cadastrar(Livros):
    
    def __init__(self, linhas):
        print("Iniciando o processo de cadastro.")
        self.__dict_Livros = {}
        self.__dict_Livros["livro"] = {}
        self.__ref = self.__dict_Livros["livro"]
        self.__y = ["y", "yes", "s", "sim"]
        self.code = "L" + str(linhas + 1)
        
    def isbn(self):
        self.__ref["isbn"] = int(input("ISBN13: "))
        isbn = self.__ref["isbn"]
        return isbn
        
    def entradas(self, parametro):
        self.__ref["title"] = input("Título: ")
        vv = input("Volume único? ")
        vv = vv.lower()
        if vv in self.__y:
            self.__ref["vol"] = "Único"
        else:
            self.__ref["vol"] = int(input("Volume: "))
        n = int(input("Quantos autores esse livro tem? "))
        if n == 1:
            self.__ref["writer"] = input("Nome do Autor: ")
        else:
            count = 0
            autor = []
            while count < n:
                a = input(f"Nome do autor {count + 1}: ")
                autor.append(a)
                count += 1
            self.__ref["writer"] = autor
        edition = input("Edição: ")
        self.__ref["ed"] = edition + "ª"
        self.__ref["year"] = int(input("Ano de publicação: "))
        self.__ref["editor"] = input("Editora: ")
        self.__ref["page"] = int(input("Número de páginas: "))
        m = int(input("Quantos gêneros esse livro tem? "))
        if m == 1:
            self.__ref["gender"] = input("Gênero: ")
        else:
            count = 0
            gen = []
            while count < m:
                g = input(f"Gênero {count + 1}: ")
                gen.append(g)
                count += 1
            self.__ref["gender"] = gen
        vi = input("O livro está no idoma original de publicação? ")
        vi = vi.lower()
        if vi in self.__y:
            idioma = input("Idoma de publicação: ")
            self.__ref["lang_o"] = idioma
            self.__ref["lang"] = "-"
            self.__ref["translate"] = "-"
        else:    
            self.__ref["lang_o"] = input("Idioma de publicação original: ")
            self.__ref["lang"] = input("Idoma de publicação: ")
            o = int(input("Quantos tradutores esse livro tem? "))
            if o == 1:
                self.__ref["translate"] = input("Tradutor(a): ")
            else:
                count = 0
                tradutores = []
                while count < o:
                    t = input(f"Tradutor(a) {count + 1}: ")
                    tradutores.append(t)
                    count += 1
                self.__ref["translate"] = tradutores 
        self.__ref["cover"] = input("Tipo de capa: ")
        if parametro == validos[0]:
            self.__ref["fornec"] = input("Quem é o Fornecedor? ")
            self.__ref["buyprice"] = float(input("Preço de compra do livro: "))
            self.__ref["profit"] = float(input("Margem de lucro: "))
            self.__ref["estoque"] = int(input("Quantidade em estoque: "))
        elif parametro == validos[1]:
            self.__ref["fornec"] = "-"
            self.__ref["buyprice"] = 0
            self.__ref["profit"] = 0
            self.__ref["estoque"] = 0
            
        # Atribuindo os valores de cada atributo    
        title = self.__ref["title"]
        vol = self.__ref["vol"]
        writer = self.__ref["writer"]
        ed = self.__ref["ed"]
        year = self.__ref["year"]
        editor = self.__ref["editor"]
        page = self.__ref["page"]
        gender = self.__ref["gender"]
        lang_o = self.__ref["lang_o"]
        lang = self.__ref["lang"]
        translate = self.__ref["translate"]
        cover = self.__ref["cover"]
        fornec = self.__ref["fornec"]
        buyprice = self.__ref["buyprice"]
        profit = self.__ref["profit"]
        estoque = self.__ref["estoque"]
        
        # Definindo o objeto
        x = Livros(title, vol, writer, ed, year, editor, isbn, page, gender, lang_o, lang, translate, cover, fornec, buyprice, profit, estoque)
        return x
    
    def cadastro(self, parametro):
        if parametro == validos[0]:
            valor = self.__ref["buyprice"] * self.__ref["profit"]
            addlivro = [self.code, self.__ref["title"], self.__ref["vol"], self.__ref["writer"], self.__ref["ed"], self.__ref["year"], self.__ref["editor"], self.__ref["isbn"], self.__ref["page"], self.__ref["gender"], self.__ref["lang_o"], self.__ref["lang"], self.__ref["translate"], self.__ref["cover"], self.__ref["fornec"], self.__ref["buyprice"], self.__ref["profit"], valor, self.__ref["estoque"]]
        elif parametro == validos[1]:
            addlivro = [self.code, self.__ref["title"], self.__ref["vol"], self.__ref["writer"], self.__ref["ed"], self.__ref["year"], self.__ref["editor"], self.__ref["isbn"], self.__ref["page"], self.__ref["gender"], self.__ref["lang_o"], self.__ref["lang"], self.__ref["translate"], self.__ref["cover"]]
        return addlivro
    
    def limpaResto(self):
        del self.__dict_Livros["livro"]
        print(f"Dicionário limpo: {self.__dict_Livros}")
```

```python
bs_or_li = input("Qual o seu tipo de negocio: ")
bs_or_li = bs_or_li.lower()
validos = ["livraria", "biblioteca"]
if bs_or_li in validos:
    
    if bs_or_li == validos[0]:
        file = "Livraria.ods"
        tabela = pd.read_excel("Livraria.ods")
    else:
        file = "Biblioteca.ods"
        tabela = pd.read_excel("Biblioteca.ods")
        
    linhas = len(tabela.index)
    cadastro = Cadastrar(linhas)
    isbn = cadastro.isbn()
    i = 0
    
    while i < linhas:
        verificaISBN = tabela["ISBN13"][i]
        if verificaISBN == isbn:
            print("Livro já adicionado!")
        else:
            pass
        i += 1
        
    livro = cadastro.entradas(bs_or_li)
    
else:
    print("Válido apenas para %ss e %ss." % (validos[0].capitalize(), validos[1].capitalize()))

addlivro = cadastro.cadastro(bs_or_li)  
tabela.loc[linhas] = addlivro
display(tabela) # Para verificar visualmente se tudo correu bem. *FUNCIONA SOMENTE EM .ipynb, OU SEJA, JUPYTER, COLAB, ETC...*

# Salvando a DB atualizada
tabela.to_excel(file, index=False)

# Limpando "restos"
cadastro.limpaResto()
```
